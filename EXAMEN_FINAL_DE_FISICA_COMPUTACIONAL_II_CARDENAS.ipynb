```
# F = m * a
# F_resorte = -k * ∆L
# F_resistencia = -q * v
```

F_resorte + F_resistencia = m * a <br> k * ∆L - q * v = m * a <BR>
la segunda derivada de la posición con respecto al tiempo <BR>
$$\frac{{d^2X}}{{dt^2}} = a$$

 $$\frac{m{d^2x}}{{dt^2}} + \frac{q{dx}}{{dt}}+ k*x = 0$$
 dy/dt = v

m * dv/dt + q * v + k * x = 0

Ahora tenemos un sistema de dos ecuaciones diferenciales de primer orden:

dy/dt = v

dv/dt = - (q/m) * v - (k/m) * x



````python

import numpy as np
import matplotlib.pyplot as plt

# Valores constantes
m = 1.0  # masa (kg)
k = 1.0  # constante elástica (N/m)
q = 0.5  # coeficiente de amortiguamiento (N*s/m)

# Condiciones iniciales
x0 = 0.0 # posición inicial (m)
v0 = 0.0  # velocidad inicial (m/s)

# Intervalo de tiempo
t0 = 0.0  # tiempo inicial (s)
tf = 10.0  # tiempo final (s)
dt = 0.01  # tamaño del paso de tiempo (s)

# Solución numérica
t = np.arange(t0, tf, dt)
x = np.zeros_like(t)
v = np.zeros_like(t)

# Condiciones iniciales
x[0] = x0
v[0] = v0

# Funciones de derivadas
def dx_dt(t, x, v):
    return v

def dv_dt(t, x, v):
    return -(q/m) * v - (k/m) * x

# Solución numérica
for i in range(1, len(t)):
    x[i] = x[i-1] + dt * dx_dt(t[i-1], x[i-1], v[i-1])
    v[i] = v[i-1] + dt * dv_dt(t[i-1], x[i-1], v[i-1])

# Gráfico de la posición en función del tiempo
plt.plot(t, x)
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')
plt.title('Posición del cuerpo colgado en el resorte')
plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del sistema
m = 1.0  # masa (kg)
k = 1.0  # constante del resorte (N/m)

# Condiciones iniciales
x0 = 0.25  # posición inicial (m)
v0 = 0.0  # velocidad inicial (m/s)

# Tiempo
t0 = 0.0  # tiempo inicial (s)
tf = 10.0  # tiempo final (s)
h = 0.01  # tamaño del paso de tiempo (s)

# Funciones de derivadas
def dx_dt(t, x, v):
    return v

def dv_dt(t, x, v):
    return -(k/m) * x

# Solución numérica utilizando Runge-Kutta de cuarto orden
t = np.arange(t0, tf, h)
x = np.zeros_like(t)
v = np.zeros_like(t)

x[0] = x0
v[0] = v0

for i in range(1, len(t)):
    k1x = h * v[i-1]
    k1v = h * dv_dt(t[i-1], x[i-1], v[i-1])

    k2x = h * (v[i-1] + k1v/2)
    k2v = h * dv_dt(t[i-1] + h/2, x[i-1] + k1x/2, v[i-1] + k1v/2)

    k3x = h * (v[i-1] + k2v/2)
    k3v = h * dv_dt(t[i-1] + h/2, x[i-1] + k2x/2, v[i-1] + k2v/2)

    k4x = h * (v[i-1] + k3v)
    k4v = h * dv_dt(t[i-1] + h, x[i-1] + k3x, v[i-1] + k3v)

    x[i] = x[i-1] + (k1x + 2*k2x + 2*k3x + k4x)/6
    v[i] = v[i-1] + (k1v + 2*k2v + 2*k3v + k4v)/6

# Gráfico de la posición en función del tiempo
plt.plot(t, x)
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')
plt.title('Posición del cuerpo colgado en el resorte (sin amortiguación)')
plt.grid(True)
plt.show()
````
