```python

import matplotlib.pyplot as plt
import numpy as np

# Rosa María Cardenas Rodríguez
# 17130077

# Intensidad de difracción
l = 480
a = 0.3
I0 = 16.0

# Rango de ángulos
x = np.linspace(0, 3*np.pi, num=100)

I = I0 * (np.sin(np.pi*a*np.sin(x)/l) / (np.pi*a*np.sin(x)/l))**2

# Gráfica
plt.plot(x, I)
plt.xlabel('θ (rad)')
plt.ylabel('I (V)')
plt.title('Intensidad de difracción')
plt.ylim(0, 16)  # Establecer límites del eje vertical
#plt.grid(True)
plt.show()
```
